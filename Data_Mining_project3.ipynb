Association Rule Mining using Apriori

```python
# Install required libraries (run once)
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# 1. Collect Transaction Dataset
transactions = [
    ['Bread', 'Milk', 'Butter'],
    ['Bread', 'Diapers', 'Beer', 'Eggs'],
    ['Milk', 'Diapers', 'Beer', 'Cola'],
    ['Bread', 'Milk', 'Diapers', 'Beer'],
    ['Bread', 'Milk', 'Diapers', 'Cola']
]

print("Transaction Data:")
for i, t in enumerate(transactions, 1):
    print(f"T{i}: {t}")
```

```python
# 2. Data Preprocessing

# Convert transactions into one-hot encoded format
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

print("\nPreprocessed Data:")
print(df)
```

```python
# 3. Apply Association Rule Mining


# Minimum support = 40%
frequent_itemsets = apriori(
    df,
    min_support=0.4,
    use_colnames=True
)

print("\nFrequent Itemsets:")
print(frequent_itemsets)
```

```python
# 4. Generate Association Rules
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6
)
```

```python
# Select important columns
rules = rules[
    ['antecedents', 'consequents',
     'support', 'confidence', 'lift']
]

print("\nAssociation Rules:")
print(rules)
```

```python
# 5. Recommendation Insights
print("\nProduct Recommendations:")

for index, row in rules.iterrows():
    antecedent = ', '.join(list(row['antecedents']))
    consequent = ', '.join(list(row['consequents']))

    print(f"If a customer buys [{antecedent}], "
          f"recommend [{consequent}] "
          f"(Confidence: {row['confidence']:.2f})")
```
